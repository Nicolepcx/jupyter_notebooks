So kann man via der Panda Data Analysis Library Exceldaten direkt einlesen, auswerten, ausgeben und/oder bereinigen. Zunnächst muss man die Panda Library importieren. Und anschliessend das zu einzulesene Excel-File.

```python
import pandas as pd                     

df = pd.read_excel("Maturiaet.xls")
df = df.fillna('')

df
```
