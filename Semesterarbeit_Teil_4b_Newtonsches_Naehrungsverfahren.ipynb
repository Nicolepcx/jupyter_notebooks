# New Section

Lineare und quadratische Gleischungen lassen sich nach einfachen algebraischen Regeln lösen. Auch bei Gleichungen dritten und vierten Grades ist das noch möglich, aber viel komplizierter. Bei algebraischen Gleichungen vom fünften Grad ab ist die Suche nach allgemeinen Lösungsformeln hoffnungslos, weil solche Lösungsformeln - wie Mathematiker nachgewiesen haben- nicht existieren. Will man daher eine Gleichung wie z. B. $x^5$+x+1=0 lösen, so ist man auf **Nährungsverfahren wie das Newton'sche Nährungsverfahren** angewiesen. Beim Newton-Verfahren wird ausgehend von einem Startwert irerativ eine Folge von Nährungswerten erzeugt, die gegen die gesuchte Sullstelle konvergiert. Dabei muss für jeden Nährungswert der Funtkionswert als auch der Wert der Ableitungsfunktion berechnet werden.

**Iterationsvorschrift des Newton-Verfahrens:** $x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$, n = 0, 1, 2,...

Möchte man zum Beispiel die Nullstelle der folgenden Funktion:
$f(x) = \frac{x^3-2x+7}{x^4+2}$

```
x = np.arange(-5,5, 0.1);
y = (x**3-2*x+7)/(x**4+2)

p1=plt.plot(x, y)
plt.xlim(-4, 4)
plt.ylim(-.5, 4)
plt.xlabel('x')
plt.axhline(0)
plt.title('Beispielfunktion: $f(x)$')
plt.show()
```

Um dieses Verfahren nutzen zu können benötigt man einen Anfang für das Interationsverfahren:

Beim Graphen lässt sich erkennen, dass die Nullstelle ca bei -2 ist, deshalb ist unsere Schätzung für die Berechnung:

$x=-1.5$

liegt kein Graph vor, legt man sich am besten eine Wertetabelle an und in dem Intervall wo der Übergang von Plus zu Minus stattfindet, liegt die Nullstelle. So lässt sich das "Ausgangs-x" für die Berechnung ermitteln. 

```
import matplotlib.pyplot as plt
import numpy as np
import scipy as scipy
from scipy.interpolate import interp1d

x = np.arange(-5,5, 0.1);
y = (x**3-2*x+7)/(x**4+2)

p1=plt.plot(x, y)
plt.xlim(-4, 4)
plt.ylim(-.5, 4)
plt.xlabel('x')
plt.axhline(0)
plt.title('Gute Schätzung mit Hilfe von Wertetabelle')
t = np.arange(-5, 5., 0.1)

x0=-1.5
xvals = []
xvals.append(x0)
notconverge = 1
count = 0
cols=['r--','b--','g--','y--','c--','m--','k--','w--']
while (notconverge==1 and count <  6):
    funval=(xvals[count]**3-2*xvals[count]+7)/(xvals[count]**4+2)
    slope=-((4*xvals[count]**3 *(7 - 2 *xvals[count] + xvals[count]**3))/(2 + xvals[count]**4)**2) + (-2 + 3 *xvals[count]**2)/(2 + xvals[count]**4)

    intercept=-slope*xvals[count]+(xvals[count]**3-2*xvals[count]+7)/(xvals[count]**4+2)

    plt.plot(t, slope*t + intercept, cols[count])
    nextval = -intercept/slope
    if abs(funval) < 0.01:
        notconverge=0
    else:
        xvals.append(nextval)
    count = count+1

plt.show()
```

Die gelbe gestrichelte Linie zeigt die Nährung der Nullstelle der Funkton $f(x)$
