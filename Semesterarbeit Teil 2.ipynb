### Graph einer Exponentialfunktion

```python
import numpy as np
import matplotlib.pyplot as plt


def main():
    x = np.linspace(-1, 2, 100)
    y = np.exp(x)

    plt.figure()
    
    plt.plot(x, y)
    plt.xlabel('$x$')
    plt.ylabel('$\exp(x)$')

  # Show graphic
    plt.show()

if __name__ == '__main__':
    main()
```

### Mehrere Graphen in einer Grafik

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

def main():
    plt.figure(figsize=(3,2))
    X = np.linspace(-np.pi, np.pi, 300)
    C,S = np.cos(X), np.sin(X)

    # Create graphs
    plt.plot([-2,-1,1,2,3,4])
    plt.plot(X,C)
    plt.plot(X,S)

    # Show graphic
    plt.show()

if __name__ == '__main__':
    main()
```

### Balkendiagramm

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

def main(): 
    # Choose the height of the bars
    height = [3, 12, 5, 18, 45]
 
    # Choose the names of the bars
    bars = ('bar1', 'bar2', 'bar3', 'bar4', 'bar5')
    y_pos = np.arange(len(bars))
 
    # Create bars
    plt.bar(y_pos, height)
 
    # Create names on the x-axis
    plt.xticks(y_pos, bars, color='green')
    plt.yticks(color='green')
 
    # Show graphic
    plt.show()
if __name__ == '__main__':
    main()
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

def main(): 
    # Fake dataset
    height = [7, 18, 5, 25, 45]
    bars = ('A', 'B', 'C', 'D', 'E')
    y_pos = np.arange(len(bars))
 
    # Create bars and choose color
    plt.bar(y_pos, height, color = (0.5,0.1,0.5,0.6))
 
    # Add title and axis names
    plt.title('fake data graph')
    plt.xlabel('categories')
    plt.ylabel('values')
 
    # Limits for the Y axis
    plt.ylim(0,60)
 
    # Create names
    plt.xticks(y_pos, bars, color ='red')
    plt.yticks(color='red')
 
    # Show graphic
    plt.show()
    
if __name__ == '__main__':
    main()
```

### Kuchendiagramm

```python
# Library
import matplotlib.pyplot as plt

def main():
    # Data to plot
    labels = 'Python', 'C++', 'Javascript', 'Java'
    sizes = [215, 100, 210, 445]
    colors = ['lightgreen', 'lightyellow', 'lightcoral', 'lightskyblue']
    explode = (0.1, 0.3, 0.1, 0.1)  # explode 1st slice
    
    # Title of graph
    plt.title('Famous Programming Languages')

    # Legend of graph
    plt.legend(labels, loc="best")
 
    # Plot
    plt.pie(sizes, explode=explode, labels=labels, colors=colors,
            autopct='%1.2f%%', shadow=True, startangle=-20)
 
    plt.axis('equal')
    
    # Show graphic
    plt.show()
    
if __name__ == '__main__':
    main()    
```

### Histogram von CSV-File

```python
# Standard import for pandas, numpy and matplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Read in the csv file 
    sales=pd.read_csv("sample-salesv2.csv",parse_dates=['date'])

    # Filter the columns 
    customers = sales[['name','ext price','date']]

    #Group the customers by name and sum their sales
    customer_group = customers.groupby('name')
    sales_totals = customer_group.sum()

    # Select the appropriate columns and group by name and category
    customers = sales[['name','category','ext price','date']]
    category_group = customers.groupby(['name','category']).sum()

    # Create a histogram of purchase volumes
    purchase_patterns = sales[['ext price','date']]
    purchase_plot = purchase_patterns['ext price'].hist(bins=20)
    purchase_plot.set_title("Purchase Patterns")
    purchase_plot.set_xlabel("Order Amount(CH)", color ='red')
    purchase_plot.set_ylabel("Number of orders", color ='green')
    
    # Show graphic
    plt.show()

if __name__ == '__main__':
    main()
```
